# グラフ

```python
# matplotlibを使ってPythonでグラフを出力する

# 散布図の描画

# Jupyterで画像作成のため以下の1行が必要
%matplotlib inline
 
import numpy as np
import matplotlib.pyplot as plt
 
# numpyで乱数を10個ずつ生成
x = np.random.rand(10)
y = np.random.rand(10)

# 値を出力
print(x,y)
 
# 散布図を描画
plt.scatter(x, y)
plt.savefig("scatter.png")
plt.show()
```

```python
# グラフのタイトル，x軸，y軸の名前 (ラベル)を追加
plt.title('scatter plot')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(x, y)
```

```python
# 折れ線グラフの描画

%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

y = np.random.rand(5)*100
print(y)

# 折れ線グラフを出力
x = np.array([1, 2, 3, 4, 5])
plt.plot(x, y)
```
